```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tabulate import tabulate

# Criteria and Weights ---
criteria = ['Sensitivity/Diesel Fuel Price ($/L)', 'System/Ren Frac (%)', 'System/Elec Prod (kWh/yr)',
            'System/CO2(kg/yr)', 'Cost/NPC ($)', 'Cost/COE ($)']
criteria

# weights add up to 1
weights = [0.333/3, 0.333/2, 0.333/2, 0.333, 0.333/3, 0.333/3]  # Approximately: [0.111, 0.1665, 0.1665, 0.333, 0.111, 0.111]

# --- 2. Calculate Appraisal Scores ---
# Extract relevant criteria columns from your DataFrame
decision_matrix = df_filled[criteria].values

decision_matrix

# Normalize the decision matrix (benefit criteria are maximized, cost criteria are minimized)
scaler = MinMaxScaler()
normalized_matrix = scaler.fit_transform(decision_matrix)

normalized_matrix

# Calculate appraisal scores (assuming a simple weighted sum method for now)
appraisal_scores = np.dot(normalized_matrix, weights)
appraisal_scores

# Rank Alternatives and Display Results ---
def rank_alternatives(appraisal_scores):
    return np.argsort(-appraisal_scores)  # Sort in descending order

rankings = rank_alternatives(appraisal_scores)
df_filled['Rank'] = np.argsort(rankings) + 1  # Add rank to DataFrame

# Display the top rankings
top_10_df = df_filled.sort_values('Rank').head(10)
print(tabulate(top_10_df[['Rank'] + criteria], headers='keys', tablefmt='psql'))


# Pick the best option (highest rank)
best_option = df_filled.loc[df_filled['Rank'] == 1]
print("\n--- Best Option ---")
best_option
```

```python
# Step 3: Construct the Decision Matrix
decision_matrix = df[['Sensitivity/Diesel Fuel Price ($/L)','Cost/NPC ($)', 'Cost/COE ($)', 'System/Ren Frac (%)','System/CO2(kg/yr)','System/Elec Prod (kWh/yr)']].values
decision_matrix
```

```python
# Step 4: Normalize the Decision Matrix
def normalize_matrix(matrix, criteria_type):
    normalized_matrix = np.zeros(matrix.shape)
    for j in range(matrix.shape[1]):
        column = matrix[:, j]
        min_val = np.min(column)
        max_val = np.max(column)
        # Adjust criteria_type to match the number of columns in decision_matrix
        if j < len(criteria_type) and criteria_type[j] == 'benefit':
            normalized_matrix[:, j] = (column - min_val) / (max_val - min_val)
        else: # Treat any column without a specified type as 'cost'
            normalized_matrix[:, j] = (max_val - column) / (max_val - min_val)
    return normalized_matrix

# Make sure criteria_type covers all columns in decision_matrix
criteria_type = ['cost', 'cost', 'benefit', 'benefit', 'cost', 'cost', 'benefit']
normalized_matrix = normalize_matrix(decision_matrix, criteria_type)

normalized_matrix
```

```python
# Step 5: Calculate PDA and NDA Matrices
def calculate_pda_nda_matrices(matrix, average_solution, criteria_type):
    pda_matrix = np.zeros(matrix.shape)
    nda_matrix = np.zeros(matrix.shape)
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            if criteria_type[j] == 'benefit':
                pda_matrix[i, j] = max(0, (matrix[i, j] - average_solution[j]) / average_solution[j])
                nda_matrix[i, j] = max(0, (average_solution[j] - matrix[i, j]) / average_solution[j])
            elif criteria_type[j] == 'cost':
                pda_matrix[i, j] = max(0, (average_solution[j] - matrix[i, j]) / average_solution[j])
                nda_matrix[i, j] = max(0, (matrix[i, j] - average_solution[j]) / average_solution[j])
    return pda_matrix, nda_matrix

average_solution = np.mean(normalized_matrix, axis=0)
pda_matrix, nda_matrix = calculate_pda_nda_matrices(normalized_matrix, average_solution, criteria_type)

pda_matrix
```

```python
nda_matrix
```

```python
# Step 6: Compute Weighted Sums
criteria_weight = np.array([0.333/3, 0.333/2, 0.333/2, 0.333, 0.333/3, 0.333/3])  # Example weights - adjust these as needed, ensuring the number of weights matches the number of columns in your matrices

def calculate_weighted_sum(matrix, weights):
    # Ensure weights are a column vector for proper broadcasting
    weights_column = weights.reshape(-1, 1)
    # Use np.dot for matrix multiplication to get weighted sums
    return np.dot(matrix, weights_column)

sp_values = calculate_weighted_sum(pda_matrix, criteria_weight)
sn_values = calculate_weighted_sum(nda_matrix, criteria_weight)

sp_values
```

```python
# Step 7: Normalize SP and SN Values
def normalize_values(values):
    return values / np.max(values)

nsp_values = normalize_values(sp_values)
nsn_values = normalize_values(sn_values)
```

```python
# Step 8: Calculate Appraisal Scores
def calculate_appraisal_score(nsp_values, nsn_values):
    return (nsp_values + nsn_values) / 2

appraisal_scores = calculate_appraisal_score(nsp_values, nsn_values)
```
